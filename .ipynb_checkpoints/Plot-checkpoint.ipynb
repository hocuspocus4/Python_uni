```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[2,3,5,11,4]

plt.plot(x,y)
plt.title("Sample")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
categories = ['Music', 'Sports', 'News', 'Opera']
values = np.random.randint(10, 100, 4)

x = np.arange(len(categories))
y = np.zeros(len(categories))   
z = np.zeros(len(categories))   

dx = np.ones(len(categories))
dy = np.ones(len(categories))
dz = values

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.bar3d(x, y, z, dx, dy, dz)

ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_ylabel('Y')
ax.set_zlabel('Popularity')
ax.set_title('Popularity of shows (3D)')

plt.show()
```

```python
categories=['Music','Sports','News','Opera']
value=np.random.randint(10,100,4)
colors=['Red','Blue','Green','Black']
z=np.arange(len(categories))

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.bar(z,value,zdir='y')
plt.bar(categories,value,color=colors)
plt.title("Popularity of shows")
plt.xticks(rotation=30)

plt.show()
```
