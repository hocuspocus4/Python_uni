```python
import tkinter as tk
root=tk.Tk()
```

```python
root.title("My first tkinter window")

tk_label=tk.Label(root,text="Welcome to tkinter")
tk_label.pack()

def on_button_click():
    tk_label.config(text="Button clicked")

click_button=tk.Button(root,text="Click",command=on_button_click)
click_button.pack()

root.mainloop()
```
